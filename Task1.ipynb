```python
img=cv2.imread('test.jpg')
```

```python
from google.colab.patches import cv2_imshow
# Display original image
cv2_imshow(img)
cv2.waitKey(0)
```

```python
# Convert to graycsale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#The function is used to blur the image which eliminates background noise. It takes grayscale image as argument
# Blur the image for better edge detection
img_blur = cv2.GaussianBlur(img_gray, (3,3), 0)
cv2_imshow(img_blur)
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Blur the image for better edge detection
kernel = np.ones((3, 3), np.float32) / 9

img = cv2.filter2D(src=img_gray, ddepth=-1, kernel=kernel)

plt.figure(figsize=(4,5))
plt.imshow(img)
```

```python
median = cv2.medianBlur(src=img_gray, ksize=5)
plt.figure(figsize=(4,5))
plt.imshow(median)
```

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

img = cv2.filter2D(src=img_gray, ddepth=-1, kernel=kernel)
plt.figure(figsize=(4,5))
plt.imshow(img)
```

```python
# Sobel Edge Detection
sobelx = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5) # Sobel Edge Detection on the X axis
sobely = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5) # Sobel Edge Detection on the Y axis
sobelxy = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5) # Combined X and Y Sobel Edge Detection
# Display Sobel Edge Detection Images
print('sobelx')
cv2_imshow(sobelx)
cv2.waitKey(0)
print('sobely')
cv2_imshow(sobely)
cv2.waitKey(0)
print('Sobel X Y using Sobel() function', )
cv2_imshow(sobelxy)
cv2.waitKey(0)
```

```python
# Canny Edge Detection
#It works based on the threshold values to eliminate the unnecessary edges 
edges = cv2.Canny(image=img_blur, threshold1=100, threshold2=200) # Canny Edge Detection
```

```python
# Display Canny Edge Detection Image
print('Canny Edge Detection')
cv2_imshow(edges)
cv2.waitKey(0)
```
